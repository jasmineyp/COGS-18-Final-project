```python
import string
import random
import nltk
```

```python
# This cell defines a collection of input and output things our chatbot can say and respond to

GREETINGS_IN = ['hello', 'hi', 'hey', 'hola', 'welcome', 'bonjour', 'greetings','howdy']
GREETINGS_OUT = ["Hello!", 'Nice to meet you!',  "Hey - Let's learn about endangered animals!", "Howdy!"]

SCIENTIFIC_NAME_IN = ['scientific', 'real', 'name', 'their']

HABITAT_IN = ['where', 'live', 'exist', 'find', 'world', 'country', 'environment', 'habitat']

ENDANGERMENT_STATUS_IN = ['status', 'level', 'endangerment', 'severity', 'how']

STATUS_REASON = ['why', 'reason', 'endangered', 'how', 'poach', 'poaching', 'poached', 'kill', 'killed', 'killing']

LAWS_IN = ['laws', 'law', 'protection', 'protect', 'government']

CHARITIES_IN = ['charities', 'charity', 'help', 'how', 'protect', 'foundation']

UNKNOWN = ['Sorry I do not have this information.', 'Come again?', 'Huh?', 'Okay!', 'Sure?', 'Yes']
```

```python
#all classes, variables, and functions

from classes import Mammals, WildCats, MarineMammals, Reptiles, Birds
from classes import bengal_tiger, amur_tiger, iberian_lynx, iriomote_cats, florida_manatee, vaquita_porpoise, mammals
from classes import hawksbill_turtle, softshelled_turtle, radiated_tortoise, reptiles
from classes import kakapo, crested_ibis, california_condor, birds

from myfunctions import is_questioin, remove_punctuation, prepare_text, selector, list_to_string, end_chat, is_in_list, find_in_list
from myfunctions import greeting_1, print_menu, print_menu2, menu_eval
```

```python
def main_chat():
    
    """The main function to run the chatbot"""
    
    interest_1 = greeting_1()
    
    #only prepare text if the user said yes to talking
    if interest_1:
        
        chat = True
        interest_1 = interest_1.lower()
        interest_1 = remove_punctuation(interest_1) 
    
        animal_choice = menu_eval(interest_1) #we set the user's choice as a variable
          
        print_menu2(animal_choice)            #print secondary menu
        animal = input("Enter the number for the one you would like to know about\nInput: ")
        animal = int(remove_punctuation(animal)) - 1   #We need int because the input is saved as a string and we need to 
                                                    # subtract one because we have the 0 marker 
    
    else: 
        chat = False 
   
                                                    
    while chat:

        # Get a message from the user
        msg = input('INPUT :\t')
        out_msg = None #initialize output as nothing

        # Check if the input is a question
        question = is_question(msg)

        # Prepare the input message
        msg = prepare_text(msg) #splits the user input into seperate words and stores into a list to be looped through

        #check if user wants to go back
        if 'back' in msg:
            main_chat()
            
        # Check for an end msg 
        if end_chat(msg):
            out_msg = 'Talk to you later!'
            chat = False
            
        
        # Check for a selection of topics that we have defined to respond to
        #   Here, we will check for a series of topics that we have designed to answer to
        if not out_msg:

            # Initialize to collect a list of possible outputs
            outs = []
            animal_name = animal_choice[animal].name

            if is_in_list(msg, GREETINGS_IN):
                
                outs.append(random.choice(GREETINGS_OUT))
                
            # Check if the input looks like a greeting, add a greeting output if so
            scientific_name = animal_choice[animal].scientific_name
            outs.append(selector(msg, SCIENTIFIC_NAME_IN, scientific_name))
            
            habitat = animal_choice[animal].habitat
            habitat = "They can be found in " + habitat
            outs.append(selector(msg, HABITAT_IN, habitat))
            
            status = animal_choice[animal].endangerment_status
            outs.append(selector(msg, ENDANGERMENT_STATUS_IN, status))
            
            reason = animal_choice[animal].reason_for_status
            reason = 'They are endangered because of ' + reason
            outs.append(selector(msg, STATUS_REASON, reason))
            
            law = animal_choice[animal].laws
            outs.append(selector(msg, LAWS_IN, law))
            
            charity = animal_choice[animal].charities
            outs.append(selector(msg, CHARITIES_IN, charity))
            
            #There is a chance we have None values, so we can't consider them as possible outputs
            return_options = list(filter(None, outs))
            if return_options:
                
                out_msg = random.choice(return_options)
        
        if not out_msg and question:
            
            animals = [0,1,2,3,4,5,6]
            animal = random.choice(animals)
            out_msg = animal_choice[animal].fun_facts
        
        if not out_msg:
            out_msg = random.choice(UNKNOWN)
        
        print('Peanut: ', out_msg)
        print('\n\nIf you would like to return to the main menu, type BACK')
```

```python
#To start a chat with Peanut
main_chat()
```
